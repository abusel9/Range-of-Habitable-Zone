The standard definition of the habitable zone is the range of distances from a star in which liquid water could exist. 

### Stellar Classification
Stars are classified into different stellar types O, B, A, F, G, K, and M. An O star is the hottest and most luminous (and subsequently has the shortest lifespan), whereas M stars are red and very dim (and can live for billions of years). These letter classifications are further subdivided into numbers 0-9, with 0 being the hottest and 9 the coldest. For example, a B3 star is hotter than a B7 star.  

### Units
- Distance from planets to their host star is measured in __AU__, astronomical units. __1AU__ is the distance from the Earth to the Sun.

- Temperature is measured in __Kelvin__ and for comparison 1K is - 457.87° F.

- Mass of stars is measured in __solar masses__, i.e. each star is given a mass that is a scalar multiple of the Sun's mass. Hence the Sun would have a mass of __1Msol (or 1M☉)__ and a star with 10 times the mass of the Sun would be said to have a mass of 10Msol (10M☉)

- Luminosity of stars is measured in __solar luminosities__. Similar concept, the sun has a luminosity of __1L☉__, and a star with a luminosity 100x that of the sun has a luminosity of 100L☉

- Radius of stars is measured in __solar radii__. An example of this would be the Sun's radius. In solar radii, it is __1Rsol (1R☉)__, and a star with 1/2 of the radius of the sun has a radius of 1/2R☉

- Mass of exoplanets is measured in __Earth masses__. One Earth's mass is __1M⊕__.

- Radius of exoplanets measured in __Earth radii, R⊕.__  

### Habitable Zones
Habitable zones are the regions around a star where liquid water could exist on an orbiting planet and has the potential to foster life. Habitable zones are calculated based on our own Sun's potential habitable zone, which is between 0.75AU and 1.77AU (with Earth located at 1AU). Of course, distance from the star isn't the only determining factor. Another aspect that can affect habitable zones would be a star's lifetime. O and B stars are very hot and luminous and therefore have very short lifespans on astronomical timescales (only a few 10Myr). Hence orbiting planets would not be able to form complex life before the star explodes. These stars are left out of consideration and are not included in our exoplanet dataset.

## Part I: Calculating Habitable Zones of Host Star

Here we are calculating each host star's luminosity from its radius and temperature. Then, we use this luminosity to find inner and outer bounds for each star's habitable zones. The functions and math to do so are shown below.


```python
# Method that calculates inner habitable zone
def inn_hz(l_star):
    sqrt_l=math.sqrt(l_star)
    r_inner=0.75*sqrt_l
    return r_inner

# Method that calculates outer habitable zone
def out_hz(l_star):
    sqrt_l=math.sqrt(l_star)
    r_outer=1.77*sqrt_l
    return r_outer

# adding new columns for stellar luminosity, and inner and outer HZ
df['Stellar Luminosity']=((df['Stellar Radius'])**2) * ((df['Stellar Effective Temperature']/5780)**4)
df['Inner Habitable Zone']=df['Stellar Luminosity'].apply(inn_hz)
df['Outer Habitable Zone']=df['Stellar Luminosity'].apply(out_hz)
df
```

## Part II: Determining Spectral Type of each Star, Letter Only
As discussed in the background section, stars are classified into a letter category depending on temperature. In our imported dataset from the NASA Exoplanet Archive, spectral types are listed very specifically. For example, G8 V (five) includes information on its spectral type (G), as well as that it is in its main sequence (V). Since we only care about the spectral letter of the spectral type, we will assign a new column that contains only the spectral letter.

However, because the spectral type is listed very specifically, and sometimes it is unknown what stage of life a star is in, some values are listed as NaN. To account for this, the effective temperature will be used to determine the spectral letter of each star.

```python
# Method to assign spectral letter depending on the stellar effective temperature
def spect_type(sp_type, teff):
    if type(sp_type) == type('a'):
        return sp_type[0]
    elif (teff > 7600 ) & (teff < 10000):
        return 'A'
    elif (teff > 6000) & (teff < 7600):
        return 'F'
    elif (teff > 5250) & (teff < 6000):
        return "G"
    elif (teff > 3900) & (teff < 5250):
        return "K"
    elif (teff <3900):
        return "M"
    else:
        return np.nan
```

```python
df['Spectral Letter']=list(map(spect_type, df['Spectral Type'], df['Stellar Effective Temperature']))
df['Spectral Letter'].value_counts()
```

## Part III: Approximating Elliptical Orbits as Circular

Orbits are in the shape of an ellipse. Due to limitations in our data as well as technology, and the fact that the star can be anywhere inside the ellipse (it does not have to be at the center), determining whether a planet is within the habitable zone of its host star becomes difficult using the ellipse model. Therefore, we would like to approximate the ellipse as a circle, and approximate that the host star is at the center of the exoplanet's orbit. To do this, we would need to show that the difference between the semi-major and semi-minor axes is minimal (close to 0). 

### Calculating Semi-Minor Axis

This will be done by using our values for the maximum planetary orbit and orbital eccentricity. We will compare this to the semi-major axis of each planet's orbit to determine if it is approximately circular.

```python
df['Orbit Semi-Minor Axis']=(((df['Orbit Semi-Major Axis'])**2)*(1-(df['Eccentricity'])**2))**(1/2)
df.head(6)
```

### Approximating The Ellipse As A Circle

Now we can check what adding the new variable means for our dataset. We can check the correlation between the semi-major and semi-minor axes to get an idea of the shape of our elliptical orbits.

```python
fig = plt.subplots(figsize = (10, 4))
u = sns.scatterplot(x = df['Orbit Semi-Minor Axis'], y = df['Orbit Semi-Major Axis'])
u.set_title("Semi-Major Axis Versus Semi-Minor Axis Measurements")
u.set_xlabel("Semi-Minor Axis Measurements")
u.set_ylabel("Semi-Major Axis Measurements")
```

There appears to be a linear relationship, but two outliers are skewing the graph, so let's view the same graph without the outliers.

Now we can see that there appears to be a linear relationship between the semi-major and semi-minor axes. We also notice that the measurements themselves for both the semi-major and semi-minor axis are concentrated around 0. This may suggest that our ellipses can be approximated by circles. We can use a T-test, but first we need to figure out whether to assume equal variance. So we start with an F-test. However, an F-test needs normality.

```python
fig2 = plt.subplots(figsize = (10, 4))
k = sns.histplot(data = df["Orbit Semi-Minor Axis"])
plt.xlim([0, 2])
plt.ylim([0, 200])
k.set_title("Distribution of Measurements of Semi-Minor Axis")
k.set_xlabel("Length of Semi-Minor Axis")
```

The distribution here is highly skewed right and not normal, so we cannot use this data for the F-test. We can check if it is possible to use a log transformation to normalize the data.

```python
#checking normality for F test
log_bins = np.logspace(base = 10, 
                       start = -3, 
                       stop = 1, 
                       num = 20) 

g = sns.histplot(
    data = df["Orbit Semi-Minor Axis"],
    bins = log_bins 
)
g.set_xscale('log')
g.set_title("Distribution of Measurements of Semi-Minor Axis")
g.set_xlabel("Length of Semi-Minor Axis")
```

(analysis for this graph included below, together with Distribution of Measurements of Semi-Major Axis)

```python
log_bins2 = np.logspace(base = 10,
                       start = -3, 
                       stop = 1, 
                       num = 20) 
p = sns.histplot(
    data = df["Orbit Semi-Major Axis"],
    bins = log_bins2 
)
p.set_xscale('log')
p.set_title("Distribution of Measurements of Semi-Major Axis")
p.set_xlabel("Length of Semi-Major Axis")
```

These distributions do appear to be roughly normal and similarly distributed. The other assumption of the F-test is that the data are independent. Since the semi-major axis does not affect the semi minor axis and since our observations are different planets, we can say that the data are independent. Let's add the log transformations to our df.

```python
df['log_min'] = np.log10(df['Orbit Semi-Minor Axis'])
df['log_max'] = np.log10(df['Orbit Semi-Major Axis'])
```

### F-test

code for F-test from https://www.statology.org/f-test-python/#:~:text=The%20F%20test%20statistic%20is%20calculated%20as%20s12,n1-1%20and%20denominator%20degrees%20of%20freedom%20%3D%20n2-1.

F-test

We will use the standard alpha level of 0.05 (95% confidence).

H0: The variances of semi-major and semi-minor axes are the same.
    
H1: The variances of semi-major and semi-minor axes are the different.

Reject H0 when pval < 0.05.

```python
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) 
    dfn = x.size-1 
    dfd = y.size-1 
    p = 1-stats.f.cdf(f, dfn, dfd)
    return f, p

print("The p-value is", f_test(df.log_min.dropna(), df.log_max.dropna())[1])
```

Our p-value can be approximated as 1. Therefore, we have very strong evidence that the variances are equal (we do not reject the null). 

### T-test
The next step is to perform the T-test, assuming equal variance. We will use the standard alpha level of 0.05 (95% confidence).

H0: The means of semi-major and semi-minor axes are equal.

H1: The means of semi-major and semi-minor axes are not equal.

Reject H0 when pval < 0.05.

```python
#t-test for means
stat, pval = stats.ttest_ind(a=df.log_min.dropna(), b=df.log_max.dropna(), equal_var=True)
print("The p-value is", pval)
```

Here, our p-value is 0.99, which is larger than our alpha level of 0.05. This means that we fail to reject the null that the means of the semi-major and semi-minor axes are equal, so we conclude that the two have equal means. As a result, we can approximate the elliptical orbit as a circle in our analysis.

## Part IV: Determining Planets in the Habitable Zone of Their Host Star

First, let's remove any entries that do not have values for the inner and outer habitable zones, as well as eccentricity.

```python
df=df[df['Inner Habitable Zone'].notnull()]
df=df[df['Outer Habitable Zone'].notnull()]
df=df[df['Orbit Semi-Major Axis'].notnull()]
df=df[df['Eccentricity'].notnull()]
df.shape
```

Now we will determine if a planet is in the habitable zone of its host star. To do this, we will check if the planet's orbit is within the inner and outer limits of their star's habitable zone. Since most of our planet's orbits can be approximated as circular, for simplicity the semi-major maxis will be used to determine whether a planet is within the habitable zone of their host star.

```python
def in_hz(orbit, inner_hab, outer_hab):
    if (orbit > inner_hab) & (orbit < outer_hab):
        in_hab_zone=True
    else:
        in_hab_zone=False
    return in_hab_zone
```

```python
df['In Habitable Zone']=list(map(in_hz, df['Orbit Semi-Major Axis'], 
                                    df['Inner Habitable Zone'], 
                                    df['Outer Habitable Zone']))
```

We will now assign a new dataset that only considers planets in the habitable zone.

## Part VI: Confirming The Above Results Through Logistic Regression

Logistic regression is a way to classify data based on certain characteristics. Here, we attempt to classify by spectral type depending on Inner and Outer Habitable Zone.

```python
X = df_in_hz[["Inner Habitable Zone", "Outer Habitable Zone"]]
y = df_in_hz["Spectral Letter"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 1)
log_reg = LogisticRegression(multi_class='ovr', solver='liblinear')
log_reg.fit(X_train, y_train)
print("The accuracy of this model is", log_reg.score(X_test, y_test))
```

We can see that the accuracy is about 59.57% which is quite low. This could be due to a certain spectral type having more range than the rest. To test this, we create multiple models, one for each spectral type. We then have the model attempt to predict whether, based on inner and outer habitable zones, a planet is or is not a specific spectral type, instead of having the model predict the spectral type, as we did above. (__NOTE__: The spectral type "A" was not included because there were no planets in the habitable zone for that spectral type in our dataset).

```python
analysiscopy = df_in_hz.copy()

analysiscopy['F'] = analysiscopy['Spectral Letter'] == 'F'
analysiscopy['G'] = analysiscopy['Spectral Letter'] == 'G'
analysiscopy['K'] = analysiscopy['Spectral Letter'] == 'K'
analysiscopy['M'] = analysiscopy['Spectral Letter'] == 'M'

analysiscopy['F'] = (analysiscopy['F']).astype(int)
analysiscopy['G'] = (analysiscopy['G']).astype(int)
analysiscopy['K'] = (analysiscopy['K']).astype(int)
analysiscopy['M'] = (analysiscopy['M']).astype(int)


for i in ['F','G','K','M']:
    x_vals = analysiscopy[['Inner Habitable Zone', "Outer Habitable Zone"]]
    y_vals = np.array(analysiscopy[i])
    xtrain, xtest, ytrain, ytest = train_test_split(x_vals, y_vals, random_state = 1)
    log_reg = LogisticRegression(random_state = 1)
    log_reg.fit(X = xtrain, y = ytrain)
    y_test_prediction = log_reg.predict(xtest)
    score = log_reg.score(X = xtest, y = ytest)
    print("Accuracy of model in predicting", i, "is ", score)
```

Here, we attempt to classify by spectral type depending on Inner and Outer Habitable Zone. Since we are using binary logistic regression, we had to create a seperate column for each spectral type, where the value is zero if the planet is not of that spectral type and 1 if it is. Each logistic regression model then classifies, based on inner habitable zone and outer habitable zone, whether the planet is of a specific spectral type. For example, our logistic regression model was around 81.36% successful in predicting if the host star of a planet was spectral type F or not (as in 81.36% of host stars were correctly predicted as either type F or not type F). 

We can see that certain spectral types have higher accuracy than others. Specifically, we can see that accuracy for the spectral type G is quite low. Does removing that spectral type from our analysis increase accuracy?

```python
df_in_hz_noG = df_in_hz[df_in_hz["Spectral Letter"] != "G"]

X = df_in_hz_noG[["Inner Habitable Zone", "Outer Habitable Zone"]]
y = df_in_hz_noG["Spectral Letter"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 1)
lm = LogisticRegression(multi_class='ovr', solver='liblinear')
lm.fit(X_train, y_train)
lm.score(X_test, y_test)
```

After completing multiple logistic regression again, this time excluding "G", the accuracy of the model has improved to 70.37%, which is quite high. Therefore, as our graphs above showed, there is a strong correlation between spectral type and habitable zone. However, certain spectral types are easier to predict than others.

The other reason is that very few planets are actually within the habitable zones of their host star (234 planets out of 4181), making up approximately 5.6% of planets in our original dataset. 

One application where our research is useful is finding planets that can be suitable for life. One can takeaway that one has more luck looking at planets with host stars that have a spectral type lower than A or F. Furthermore, one can use the relationships between solar mass and radius to narrow down planets that are within the habitable zone.
